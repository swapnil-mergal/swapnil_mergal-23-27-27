Q. A] Write a Python program using NumPy to create a random 5x5 matrix with values ranging
between 1 and 100. Then perform the following tasks:
1. Replace all even numbers in the matrix with -1.
2. Calculate the mean value of each row and store these values in a new array.
3. Normalize the matrix such that the values lie between 0 and 1 using min-max
normalization.

Normalization formula: xnorm=(x−min)/(max−min)

```python
import numpy as np
# Create a random 5x5 matrix
matrix = np.random.randint(1, 101, size=(5, 5))
```

```python
# Replace all even numbers in matrix with -1
matrix[matrix % 2 == 0] = -1
```

```python
# Calculate the mean value of each row and store these values in a new array
row_means = np.mean(matrix, axis=1)
```

```python
# Normalize the matrix such that the values lie between 0 and 1 using min-max normalization.
min_val = np.min(matrix)
max_val = np.max(matrix)
normalized_matrix = (matrix - min_val) / (max_val - min_val)
print("Modified Matrix:\n", matrix)
print("Row Means:\n", row_means)
print("Normalized Matrix:\n", normalized_matrix)
```

B]. Create a Pandas DataFrame containing 1000 rows and 5 columns, filled with random
integers between 1 and 500. Perform the following operations:
1. Calculate the 50th percentile for each column.
2. Use groupby on the columns to categorize values as "Low", "Medium", and
"High" based on their value ranges (1-150: Low, 151-350: Medium, 351-500:
High).
3. Find the sum of each category after grouping and plot a bar graph to display the
results.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame with random integers
df = pd.DataFrame(np.random.randint(1, 501, size=(1000, 5)))
# calculate 80th percentile for each column
percentiles = df.quantile(0.5)
# Use groupby on the columns to categorize values as "Low", "Medium", and "High" based on their value ranges (1-150: Low, 151-350: Medium, 351-500: High).
df['category'] = pd.cut(df[0], bins=[0, 150, 350, 500], labels=['Low', 'Medium', 'High'])
# Find the sum of each category after grouping and plot a bar graph to display the results
category_sums = df.groupby('category').sum()
category_sums.plot(kind='bar')
plt.show()
```

C] Write a Python program to:
1. Create a 3D NumPy array of shape (3, 4, 5) with random values between 1 and
50.
Compute the mean value along axis 2 and store it in a new 2D array.

3.Transpose the new array and then flatten it into a 1D array.

4. Plot a scatter plot using the indices as x-values and the flattened values as yvalues.

Hint: Use np.mean(), np.transpose(), and np.flatten() functions

```python
#Create a 3D NumPy array of shape (3, 4, 5) with random values between 1 and 50
array_3d = np.random.randint(1, 51, size=(3, 4, 5))
# Compute the mean value along axis 2 and store it in a new 2D array.
mean_values = np.mean(array_3d, axis=2)
# Transpose the new array and then flatten it into a 1D array.
transposed_array = np.transpose(mean_values)
flattened_array = transposed_array.flatten()
# Plot a scatter plot using the indices as x-values and the flattened values as yvalues.
plt.scatter(np.arange(len(flattened_array)), flattened_array)
plt.title('Scatter Plot of Flattened Values')
plt.xlabel('Index')
plt.ylabel('Flattened Values')
plt.show()
```

D] Given a large dataset sales.csv containing the columns Date, Product, Region, Sales,
and Cost, write a Python program using Pandas to:
1. Create a pivot table with Product as rows, Region as columns, and Sales as
values.
2. Add a new column to the pivot table for each region showing profit (Sales - Cost).
3. Plot a heatmap of the profit for each product-region combination

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Date': pd.date_range(start='1/1/2023', periods=100),
    'Product': ['Product A', 'Product B', 'Product C'] * 33 + ['Product A'],
    'Region': ['North', 'South', 'East'] * 33 + ['North'],
    'Sales': [1000, 1500, 800] * 33 + [900],
    'Cost': [600, 900, 500] * 33 + [550]
}

df = pd.DataFrame(data)
df.to_csv('sales.csv', index=False)

# Read sales.csv
df = pd.read_csv('sales.csv')

# Create a pivot table
pivot_table = df.pivot_table(index='Product', columns='Region', values='Sales')

# Add profit column for each region
pivot_table['Profit'] = df['Sales'] - df['Cost']

# Plot a heatmap of profit
sns.heatmap(pivot_table)
plt.title("Profit Heatmap")
plt.show()

```

# 5. Write a Python program to generate 1000 random points representing (x, y) coordinates
using np.random.rand(). Use matplotlib to:
1. Plot a scatter plot with varying marker size depending on y values.
2. Apply color mapping based on x values.
3. Customize the plot with labels for x-axis, y-axis, and a legend indicating the size
and color mapping.
Hint: Use plt.scatter() with c and s parameters

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random (x, y) points
x = np.random.rand(1000)
y = np.random.rand(1000)

# Scatter plot with varying marker size and color mapping
plt.scatter(x, y, c=x, s=y*100, cmap='viridis', alpha=0.6)
plt.colorbar(label='X values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Size and Color Mapping')
plt.show()
```
